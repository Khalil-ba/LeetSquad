```python
import huggingface_hub as hf_hub
from datasets import load_dataset
import pandas as pd
```

```python
# Login using e.g. `huggingface-cli login` to access this dataset
hf_hub.login()

ds = load_dataset("newfacade/LeetCodeDataset", split="train")

print(ds)
```

```python
df = pd.DataFrame(ds)
```

```python
df.head()
```

```python
import os

df_full = pd.DataFrame(ds)  

test_cases_dir = "../test_cases"
os.makedirs(test_cases_dir, exist_ok=True)

unique_prompts = df_full['prompt'].unique()

utils_py_path = os.path.join(test_cases_dir, "utils.py")

with open(utils_py_path, 'w', encoding='utf-8') as f:
    f.write("# Combined LeetCode Problem Utils\n")
    f.write("# Auto-generated from all unique prompts in the dataset\n\n")
    
    # Write all unique prompts (which contain the utility code)
    for i, prompt in enumerate(unique_prompts):
        if prompt and str(prompt).strip():  # Skip empty prompts
            f.write(f"# === Prompt {i+1} ===\n")
            f.write(str(prompt))
            f.write("\n\n")

print(f"Created combined utils.py at: {os.path.abspath(utils_py_path)}")
print(f"Combined {len(unique_prompts)} unique prompts")
```

```python
os.makedirs(test_cases_dir, exist_ok=True)

for index, row in df_full.iterrows():
    task_id = row['task_id']
    test_code = row['test']
    
    filename = f"{task_id}.py"
    filepath = os.path.join(test_cases_dir, filename)
    
    with open(filepath, 'w', encoding='utf-8') as f:
        f.write("# Import the utils module for prompts\n")
        f.write("from utils import *\n\n")
        f.write(test_code)
    
    print(f"Created: {filepath}")

print(f"\nTotal test files created: {len(df)}")
print(f"Files saved in: {os.path.abspath(test_cases_dir)}")
```
